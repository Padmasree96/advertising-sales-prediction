```python
import numpy as np
import matplotlib.pyplot as plt  
import seaborn as sns
import pandas as pd  
data_set = pd.read_csv("Advertising Budget and Sales.csv")
data_set
```

```python
x = data_set[["TV Ad Budget ($)","Radio Ad Budget ($)","Newspaper Ad Budget ($)"]]
x
```

```python
y = data_set[["Sales ($)"]]
y
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(x,y, test_size = 0.2, random_state = 42)
```

```python
x_train
```

```python
y_train
```

```python
x_test
```

```python
y_test
```

```python
from sklearn.linear_model import LinearRegression
```

```python
model= LinearRegression()  
model.fit(x_train, y_train)  
```

```python
y_pred = model.predict(x_test)
x_pred = model.predict(x_train)
```

```python
y_pred
```

```python
from sklearn.metrics import r2_score, mean_squared_error

y_test = np.ravel(y_test)
y_pred = np.ravel(y_pred)
```

```python
r2 = r2_score(y_test, y_pred)
rmse = np.sqrt(mean_squared_error(y_test, y_pred))
```

```python
plt.figure(figsize=(6, 6))
sns.scatterplot(x=y_test, y=y_pred, alpha=0.6, color="black")
plt.plot([y_test.min(), y_test.max()], [y_test.min(), y_test.max()], "--r", label="Perfect prediction")
plt.xlabel("Actual Sales ($)")
plt.ylabel("Predicted Sales ($)")
plt.title("Actual vs Predicted Sales")
plt.text(y_test.min(), y_test.max() * 0.9, f"R² = {r2:.2f}\nRMSE = {rmse:.2f}")
plt.grid(True, linestyle="--", alpha=0.5)
plt.axis("equal") 
plt.legend()
plt.show()
```
